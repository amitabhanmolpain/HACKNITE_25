```python
import pandas as pd

df = pd.read_csv("NCERT_dataset.csv")
```

```python
print(df.info())      # Overview of columns and data types

```

```python
print(df.head())      # First few rows
print(df.isnull().sum())  # Count missing values
```

```python
print(df.columns)  # Checks if there's a column related to subjects
```

```python
df['Difficulty'].unique()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Question'] = df['Question'].str.lower()
df['Answer'] = df['Answer'].str.lower()
df['subject'] = df['subject'].str.lower()
df['Topic'] = df['Topic'].str.lower()
```

```python
import re

def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
        text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

df['Question'] = df['Question'].apply(clean_text)
df['Answer'] = df['Answer'].apply(clean_text)
df['Topic'] = df['Topic'].apply(clean_text)
df['subject'] = df['subject'].apply(clean_text)
```

```python
df['Difficulty'] = pd.Categorical(df['Difficulty'], categories=['easy', 'medium', 'hard'], ordered=True)
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
